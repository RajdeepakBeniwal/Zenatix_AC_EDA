```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df=pd.read_csv('AC_Data.csv')
df.head()
```

```python
df.info()
```

```python
df['0']=pd.to_datetime(df['0'])
```

```python
df.set_index('0', inplace=True)
```

```python
df=df.replace('NaN','NA')
df1=df.dropna(how='all')    #dropping all rows with Nan value
```

```python
df1.head()
```

```python
column=['AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7', 'AC 8', 'AC 9',
       'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15', 'AC 16', 'AC 17',
       'AC 18']
```

```python
for i in column:
    print(f'{i}---------------\n{df1[i].isnull().value_counts()}')
```

```python
df1['Month']=df1['0'].dt.month
df1['Day']=df1['0'].dt.day
df1['Hour']=df1['0'].dt.hour
```

```python
df_aug=df1[:29872]    # separate dataframe for august month
df_sep=df1[29872:]    # separate dataframe for september month
```

```python
# total consumption over 2 months
total_consumption_dict={}
for i in column:
    print(f'{i}--------------{df[i].sum()}')
    total_consumption_dict[i]=df[i].sum()
```

```python
plt.figure(figsize=(15,5))
plt.bar(total_consumption_dict.keys(),total_consumption_dict.values(),width=.5,color='g')
```

Here maximium power is consumed by AC 18 and Minimum by AC 5 over a period of 2 months

```python
# maximum consumption in 1 day
max_consumption_dict={}
for i in column:
    print(f'{i}--------------{df[i].max()}')
    max_consumption_dict[i]=df[i].max()
```

```python
plt.figure(figsize=(15,5))
plt.bar(max_consumption_dict.keys(),max_consumption_dict.values(),width=.5,color='r')
```

Maximum power consumed by each AC in one day,

Here maximium power is consumed by AC 18 and Minimum by AC 5 over a period of 2 months
